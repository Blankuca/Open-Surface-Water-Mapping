# Noise filtering

```python
path = "H:/sentinel2/Ghana/Validation/02-01-20/SE/Obom.tif"
root = 'H:/sentinel2/Ghana/Validation/02-01-20/SE/Shapefiles/'

t = Image.make_dataset(path)
X = t['df']
dim = t['dim']

y_pred = LogReg.predict(X)
y_pred = np.array([1 if x<3 else 0 for x in y_pred])
y_pred = ndimage.median_filter(y_pred.reshape(dim), size=2)
Image.save_as_tiff('temp.tif', path, y_pred)

y_true = Processing.make_dataset(root, path).Content
y_true = np.array([1 if x<3 else 0 for x in y_true])

y_pred = Validation.make_dataset2(root, 'temp.tif').Content
#y_pred, _ = Validation.get_ypred(root, 'temp.tif', y_true)
```

```python
Validation.evaluate(y_true,y_pred,print_cm = True, print_err=True);
```

```python
Validation.binary(y_true, y_pred)
```
